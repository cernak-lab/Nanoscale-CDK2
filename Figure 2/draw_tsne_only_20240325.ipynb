```python
import dill

data = dill.load(open("all_data_for_tsne.pkl", "rb"))
Z, n, n2, marked = data["Z"], data["n"], data["n2"], data["marked"]
```

```python
import matplotlib.pyplot as plt


fig,ax = plt.subplots(dpi=300, figsize=(3,3))
ax.scatter(Z[0:n,0],Z[0:n,1], c="grey", s=3, alpha=0.01)
ax.scatter(Z[n:n2,0],Z[n:n2,1], zorder=2, c="black", s=1, alpha=1)
ax.scatter(Z[n2:,0],Z[n2:,1], zorder=3, c="black", s=1, marker='x')
# ax.scatter(marked[:,0],marked[:,1], zorder=4, marker="x",c="black", s=5)

ax.set_title("tested vs enumerated amide products", fontsize=6, fontfamily="Arial")
ax.set_xlabel("tSNE1", fontsize=6, fontfamily="Arial")
ax.set_ylabel("tSNE2", fontsize=6, fontfamily="Arial")

ax.set_xticks([])
ax.set_yticks([])
ax.set_xticklabels([])
ax.set_yticklabels([])
# plt.tight_layout()

plt.savefig("amide_products_overlaid_tsne_both_plates_with_unobtained.png", dpi=300, bbox_inches="tight", pad_inches=0.01)
```
