```python
from rdkit import Chem
import pandas as pd


mols = pd.read_excel("enum_reactions_amide.xlsx") # comm av acids i guess

clean_acids_all = []
for k in mols["Reactant 2"]:
    mol = Chem.MolFromSmarts(k)
    try:
        Chem.SanitizeMol(mol)
    except:
        continue
    sm = Chem.MolToSmiles(mol)
    if sm not in clean_acids_all:
        clean_acids_all.append(sm)
print(len(clean_acids_all))
```

```python
tested = pd.read_csv("james1536_input_soluble.csv")
tested_acids = []
for i,k in enumerate(tested["chemtype"]):
    if k == "Electrophile":
        tested_acids.append(tested["smiles"][i])
amide1_tested_acid_sm = []
for k in tested_acids:
    mol = Chem.MolFromSmiles(k)
    Chem.SanitizeMol(mol)
    sm = Chem.MolToSmiles(mol)
    amide1_tested_acid_sm.append(sm)
print(len(amide1_tested_acid_sm))
```

```python
tested = pd.read_csv("james1536_input_soluble_plate2.csv")
tested_acids = []
for i,k in enumerate(tested["chemtype"]):
    if k == "Electrophile":
        tested_acids.append(tested["smiles"][i])
amide2_tested_acid_sm = []
for k in tested_acids:
    mol = Chem.MolFromSmiles(k)
    Chem.SanitizeMol(mol)
    sm = Chem.MolToSmiles(mol)
    amide2_tested_acid_sm.append(sm)
print(len(amide2_tested_acid_sm))
```

```python
from rdkit.Chem import AllChem
rxn = AllChem.ReactionFromSmarts("[CX3:1](=O)[OX2H1].[NX3;H2,H1;!$(NC=O):2]>>[CX3:1](=O)[NX3;H1;!$(NC=O):2]")
amines = ["CC1(NC(O[C@@H]2CC[C@H](C3=CC(N)=NN3)C2)=O)CC1", "NC1=NNC([C@@H]2C[C@H](OC(NC(C)C)=O)CC2)=C1", "NC1=NNC([C@@H]2C[C@H](OC(N(C)C(C)C)=O)CC2)=C1", \
          "C[C@H](C1CC1)NC(O[C@@H]2CC[C@H](C3=CC(N)=NN3)C2)=O", "O=C(O[C@@H]1CC[C@H](C2=CC(N)=NN2)C1)N3CCC3(C)C", "O=C(O[C@@H]1CC[C@H](C2=CC(N)=NN2)C1)N3[C@@H](C)CC[C@@H]3C"]
am_mols = [Chem.MolFromSmiles(k) for k in amines]
```

```python
plate1_acids = [Chem.MolFromSmiles(k) for k in amide1_tested_acid_sm]
plate2_acids = [Chem.MolFromSmiles(k) for k in amide2_tested_acid_sm]
all_acids = [Chem.MolFromSmiles(k) for k in clean_acids_all]
```

```python
print(len(plate1_acids), len(plate2_acids), len(all_acids))
```

```python
amide1_prods = []
amide2_prods = []
for k in plate1_acids:
    for j in am_mols:
        p = rxn.RunReactants((k,j))[0][0]
        Chem.SanitizeMol(p)
        amide1_prods.append(p)
for k in plate2_acids:
    for j in am_mols:
        p = rxn.RunReactants((k,j))[0][0]
        Chem.SanitizeMol(p)
        amide2_prods.append(p)
print(len(amide1_prods),len(amide2_prods))
```

```python
all_enumerated_prods = []

for k in all_acids:
    if not k:
        continue
    for j in am_mols:
        p = rxn.RunReactants((k,j))[0][0]
        if not p:
            continue
        Chem.SanitizeMol(p)
        all_enumerated_prods.append(p)
```

```python
print(len(all_enumerated_prods))
```

```python
import numpy as np
from rdkit.Chem import DataStructs


all_amides_fps = []
amide_label = []
all_smiles = []
n = 0
n2 = 0
for mol in all_enumerated_prods:
    fp = np.zeros((1,))
    DataStructs.ConvertToNumpyArray(AllChem.GetMorganFingerprintAsBitVect(mol, 4,nBits=2048), fp)
    all_amides_fps.append(fp)
    amide_label.append(0)
    all_smiles.append(Chem.MolToSmiles(mol))
    n = n + 1
    n2 = n2 + 1
for mol in amide1_prods:
    fp = np.zeros((1,))
    DataStructs.ConvertToNumpyArray(AllChem.GetMorganFingerprintAsBitVect(mol, 4,nBits=2048), fp)
    all_amides_fps.append(fp)
    amide_label.append(1)
    all_smiles.append(Chem.MolToSmiles(mol))
    n2 = n2 + 1
for mol in amide2_prods:
    fp = np.zeros((1,))
    DataStructs.ConvertToNumpyArray(AllChem.GetMorganFingerprintAsBitVect(mol, 4,nBits=2048), fp)
    all_amides_fps.append(fp)
    amide_label.append(2)
    all_smiles.append(Chem.MolToSmiles(mol))
```

```python
not_obtained = pd.read_excel("unobtained.xlsx")
```

```python
canon_sm = []

for k in not_obtained["smiles"]:
    canon_sm.append(Chem.MolToSmiles(Chem.MolFromSmiles(k)))
print(len(canon_sm))
```

```python
cnt = 0

for k in all_smiles:
    if k in canon_sm:
        cnt = cnt + 1
print(cnt)
```
